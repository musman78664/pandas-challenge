```python
# Total Schools
total_schools = school_data['school_name'].count()
print('Total Schools: ', total_schools)
```

```python
# Total Students
total_students = student_data['Student ID'].count()
print('Total Students: ', total_students)
```

```python
# Total Budget
total_budget = school_data['budget'].sum()
print('Total Budget: ', total_budget)
```

```python
# Average Math Score
average_math_score = student_data['math_score'].mean()
print('Average Math Score: ', average_math_score)
```

```python
# Average Reading Score
average_reading_score = student_data['reading_score'].mean()
print('Average Reading Score: ', average_reading_score)
```

```python
# % Passing Math
passing_math = student_data.loc[student_data['math_score'] >= 70, ['math_score']]
passing_math.head()
```

```python
# number of student passing math
passing_math_count = passing_math['math_score'].count()
print(passing_math_count)
```

```python
# Passing Math score, had to multiplying by 100 to get the percentage
percent_passing_math = passing_math_count / total_students * 100
percent_passing_math
```

```python
# % Passing Reading
passing_reading = student_data.loc[student_data['reading_score'] >= 70, ['reading_score']]
passing_reading.head()
```

```python
# number of student passing Reading
passing_reading_count = passing_reading['reading_score'].count()
print(passing_reading_count)
```

```python
# Passing Reading score, had to multiplying by 100 to get the percentage
percent_passing_reading = passing_reading_count/total_students*100
percent_passing_reading
```

```python
# Overall Passing Rate (Average of the above two)
overall_passing = (percent_passing_reading + percent_passing_math)/2 
overall_passing
```

```python
# District Summary
district_summary = pd.DataFrame({"Total Schools": [total_schools],
                                "Total Students": [total_students],
                                "Total Budget": [total_budget],
                                "Average Math Score": [average_math_score],
                                "Average Reading Score": [average_reading_score],
                                "% Passing Math": [percent_passing_math],
                                "% Passing Reading": [percent_passing_reading],
                                "% Overall Passing Rate": [overall_passing],
                                })

# Organize the columns
district_summary = district_summary[["Total Schools", 
                                     "Total Students", 
                                     "Total Budget", 
                                     "Average Math Score", 
                                     "Average Reading Score", 
                                     "% Passing Math", 
                                     "% Passing Reading", 
                                     "% Overall Passing Rate"
                                    ]]


# Format to Total Budget to include $ sign as $12345.00
district_summary['Total Budget'] = district_summary['Total Budget'].map('${:,.2f}'.format)
district_summary         
```
